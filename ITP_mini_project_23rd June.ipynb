### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

**2. Rectangle**<br>

**Please enter the length of a side: _**

**Invalid input, please enter a number: _**


```python
## Area of Square
def aosq(side):
    areas=side*side
    return areas

# Area of Rectangle
def aorect(leng,bread):
    arear=leng*bread
    return arear

# Area of Triangle
def aotri(base,heigh):
    areat=(1/2)*base*heigh
    return areat

# Area of Circle
def aocir(rad):
    pi=3.142
    areac=pi*(rad*rad)
    return areac

# Question.
ques=("Which shape would you like to calculate the area for?")
print(ques)
list1=['1. Square','2. Rectangle','3. Triangle','4. Circle']
for i in list1:
    print(i)
# Selecting the input from user.
try:
    choose=int(input("Please select the number of the Shape: "))
    if choose==1:
            side=int(input("Enter the Value of Side of the Square: "))
            print("The Area of Square is: %s"%(aosq(side)))
    elif choose==2:
            leng=int(input("Enter the value of Length of Rectangle: "))
            bread=int(input("Enter the value of Breadth of Rectangle: "))
            print("The Area of Reactangle is: %s"%(aorect(leng,bread)))
    elif choose==3:
            base=int(input("Enter the value of Base of the Triangle: "))
            heigh=int(input("Enter the value of Height of the Triangle: "))
            print("The Area of Triangle is: %s"%(aotri(base,heigh)))
    elif choose==4:
            rad=int(input("Enter the value of Radius of the Circle: "))
            print("The Area of Cirlce is: %s"%(aocir(rad)))
    else:
        print("Choose from the given list!")
except:
    print("Invalid input,Please enter a valid input!")
```

### Problem 2:

Create and encrypter in python based on the ceaser cipher. It is a substitution cipher where each character of the original text is shifted a certain number characters in the alphabet. Write a function that would require 2 arguments – the input text to be encrypted and a key. For eg: Given the input text ‘hello’ and the key 3, the resulting encryted text would be ‘khoor’. Here you can see that every character in the string hello is shifted by 3 characters. ‘h’ has shifted to ‘k’, ‘e’ has shifted to ‘h’ and so on. If a key of 5 were used, the resulting string would be ‘mjqqt’. This function should be capable of ignoring any characters which are not alphabets. The character ‘z’ entered by the user for a key of 3 would result in ‘c’.
